```python
# --- read data ---
idir = '/g/data/p66/ars599/CMIP6/APP_output/CMIP6/CMIP/CSIRO-ARCCSS/ACCESS-CM2/piControl/'
r1Name = 'r1i1p2f1/Amon/rsdt/gn/v20230616/rsdt_Amon_ACCESS-CM2_piControl_r1i1p2f1_gn_050001-059912.nc'
r2Name = 'r1i1p2f1/Amon/rsut/gn/v20230616/rsut_Amon_ACCESS-CM2_piControl_r1i1p2f1_gn_050001-059912.nc'
r3Name = 'r1i1p2f1/Amon/rlut/gn/v20230616/rlut_Amon_ACCESS-CM2_piControl_r1i1p2f1_gn_050001-059912.nc'

# Open a netCDF data file using xarray default engine and load the data into xarrays
r1_in = xr.open_dataset(idir+r1Name)
r2_in = xr.open_dataset(idir+r2Name)
r3_in = xr.open_dataset(idir+r3Name)

# Use date as the dimension rather than time
year_start = 500
year_end = 599

r1_ds = r1_in.sel(time=slice(f'{year_start:04d}', f'{year_end:04d}')) # 192 x 145
r2_ds = r2_in.sel(time=slice(f'{year_start:04d}', f'{year_end:04d}')) # 192 x 145
r3_ds = r3_in.sel(time=slice(f'{year_start:04d}', f'{year_end:04d}')) # 192 x 144
```

```python
rsdt = r1_ds['rsdt'].mean(['lat','lon'])
rsut = r2_ds['rsut'].mean(['lat','lon'])
rlut = r3_ds['rlut'].mean(['lat','lon'])
qnett_r2 = rsdt - rsut - rlut
```

```python
# --- read data ---
idir = '/g/data/p66/ars599/CMIP6/APP_output/CMIP6/CMIP/CSIRO-ARCCSS/ACCESS-CM2/piControl/'
r1Name = 'r1i1p3f1/Amon/rsdt/gn/v20230517/rsdt_Amon_ACCESS-CM2_piControl_r1i1p3f1_gn_050001-059912.nc'
r2Name = 'r1i1p3f1/Amon/rsut/gn/v20230517/rsut_Amon_ACCESS-CM2_piControl_r1i1p3f1_gn_050001-059912.nc'
r3Name = 'r1i1p3f1/Amon/rlut/gn/v20230517/rlut_Amon_ACCESS-CM2_piControl_r1i1p3f1_gn_050001-059912.nc'

# Open a netCDF data file using xarray default engine and load the data into xarrays
r1_in = xr.open_dataset(idir+r1Name)
r2_in = xr.open_dataset(idir+r2Name)
r3_in = xr.open_dataset(idir+r3Name)

# Use date as the dimension rather than time
year_start = 500
year_end = 599

r1_ds = r1_in.sel(time=slice(f'{year_start:04d}', f'{year_end:04d}')) # 192 x 145
r2_ds = r2_in.sel(time=slice(f'{year_start:04d}', f'{year_end:04d}')) # 192 x 145
r3_ds = r3_in.sel(time=slice(f'{year_start:04d}', f'{year_end:04d}')) # 192 x 144
```

```python
rsdt = r1_ds['rsdt'].mean(['lat','lon'])
rsut = r2_ds['rsut'].mean(['lat','lon'])
rlut = r3_ds['rlut'].mean(['lat','lon'])
qnett_r3 = rsdt - rsut - rlut
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate the 12-month running mean of crf
qnett2_rolling = qnett_r2.rolling(time=12, center=True).mean()
qnett3_rolling = qnett_r3.rolling(time=12, center=True).mean()

# Plot the time series
plt.plot(qnett_r2.time, qnett2_rolling,label='R2')
plt.plot(qnett_r3.time, qnett3_rolling,label='R3')

# Set the x and y labels
plt.xlabel('Time')
plt.ylabel('QNET (TOA)')

# Add a grid
plt.grid(True)

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
