# historical repeat

```python
# --- read data ---
idir = '/scratch/p66/ars599/archive/'
v1Name = 'cw787/history/ocn/ocean_scalar.nc-*'
v2Name = 'cx614/history/ocn/ocean_scalar.nc-*'
v3Name = 'cx923/history/ocn/ocean_scalar.nc-*'

# Open a netCDF data file using xarray default engine and load the data into xarrays
f1name = idir+v1Name
f2name = idir+v2Name
f3name = idir+v3Name
ds1 = xr.open_mfdataset(f1name, combine='by_coords')
ds2 = xr.open_mfdataset(f2name, combine='by_coords')
ds3 = xr.open_mfdataset(f3name, combine='by_coords')

# Use date as the dimension rather than time
#year_start = 550
#year_end = 599
#df = ds.sel(time=(ds['time.year'] >= year_start) & (ds['time.year'] <= year_end)) # 192 x 144
```

```python
# Create a figure and axes with custom size
fig, ax = plt.subplots(figsize=(10, 6))

plt.plot(ds1.time, ds1.ke_tot, label='cw787')
plt.plot(ds2.time, ds2.ke_tot, label='cx614')
plt.plot(ds3.time, ds3.ke_tot, label='cx923')

plt.ylabel('Energy (10^15 Joules)')
plt.xlabel('Year')
plt.grid()
plt.legend()

plt.savefig('ke_tot_pdhist_repeat.png')
```
